# Data Preprocessing

## Difference Between StandardScaler and RobustScaler

- **StandardScaler**: Scales features to have a mean of 0 and standard deviation of 1.  
  **Formula**: \((X - \text{mean}) / \text{std}\)  
  **When to use**: Suitable for normally distributed data or algorithms sensitive to feature scaling (e.g., SVM, PCA).  
  **Limitation**: Sensitive to outliers.

- **RobustScaler**: Scales features using the median (centers at 0) and IQR (interquartile range).  
  **Formula**: \((X - \text{median}) / \text{IQR}\)  
  **When to use**: Works well for data with outliers or non-uniform distributions.  
  **Limitation**: May not perform well on normally distributed data.

**Recommendation**: Use StandardScaler for normal distributions, RobustScaler for datasets with outliers.

```python
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()
scaled_data = scaler.fit_transform(train)
scaled_data
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=3, random_state=42)
pca_components = pca.fit_transform(scaled_data)
pca_df = pd.DataFrame(pca_components, columns=['PC1','PC2','PC3'])
pca_df
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(pca_df['PC1'], pca_df['PC2'],pca_df['PC3'])
plt.show()
```

# Modelling and selecting the best k using Elbow Method

```python
from sklearn.cluster import KMeans
fit_models = {}

for i in range(2,21):
    model = KMeans(n_clusters=i, n_init=10)
    model.fit(train)
    fit_models[i] = model
```

```python
results = []
for k, model in fit_models.items():
    res = model.inertia_
    results.append(res)
```

```python
plt.title('Sum of Squared Errors aka Inertia')
plt.plot(range(2,21), results)
plt.show()
```

```python
plt.title('Sum of Squared Errors aka Inertia')
plt.plot(range(2,8), results[:6])
plt.show()
```
